### ¬øCon qu√© frecuencia los usuarios regresan?

```python
# A√±adir campos de primera visita, primera semana y primer mes. 
# Primera visita
first_visit_date = visits_v2.groupby('uid')['visit_date'].min()
first_visit_date.name =  'first_visit_date'
visits_v2 = visits_v2.join(first_visit_date, on='uid')
visits_v2.head()
```

```python
# A√±adir la primera semana registrada de visita
visits_v2['first_visit_week'] = pd.to_datetime(visits_v2['first_visit_date'], unit='d') - pd.to_timedelta(visits_v2['first_visit_date'].dt.dayofweek, unit='d')
visits_v2
```

```python
# A√±adir el mes de visita y el primer mes registrado
visits_v2['first_visit_month'] = visits_v2['first_visit_date'].astype('datetime64[M]')
visits_v2.head()
```

```python
#Calculamos los d√≠as transcurridos entre la primera sesi√≥n
visits_v2['return_days'] = visits_v2['visit_month'] - visits_v2['first_visit_month']
visits_v2
```

```python
# Excluimos las visitas realizadas el mismo d√≠a
visits_v2[visits_v2['return_days'] != '0 days']
```

```python
# Ahora verificamos visitas realizadas el mismo d√≠a
visits_v2[visits_v2['return_days'] == '0 days']
```

<b>Comentarios:</b> Excluimos del √°nalisis los registros correspondiente a cero d√≠as de visita, ya que los que nos interesa verificar el plazo transcurrido para que un usuario regrese. 

```python
users_return_days = visits_v2[visits_v2['return_days'] != '0 days'].groupby('uid').agg({'return_days':'min', 'visit_date':'count'}).reset_index()
users_return_days['return_days'] = users_return_days['return_days'].dt.days
users_return_days.head()
```

La siguiente agrupaci√≥n nos permitir√° contar el n√∫mero de usuarios que vuelven a al primer, segundo, tercer d√≠a, etc.

```python
return_count = users_return_days.groupby('visit_date')['uid'].count().reset_index()
return_count.head(30)
```

Generamos un histograma de d√≠as trasncurridos para el retorno de un usuario. Observamos que la mayor√≠a de los usuarios retorna dentro de los primeros 50 d√≠as posteriores a la primera visita. 

```python
users_return_days['return_days'].hist(bins=20)
plt.show()
```

La siguiente distribuci√≥n nos muestra el n√∫mero de visitas que cada usuario registra durante un a√±o. Observamos que la mayor parte de los usuarios realiza pocas visitas

```python
return_count['visit_date'].hist(bins=30)
plt.show()
```

Ahora verificaremos el comportamiento de retorno de visita por cohortes mensuales.

```python
#Analisis de retorno de visitas y actividad por cohortes mensuales. 
visits_report_1 = visits_v2[visits_v2['return_days'] != '0 days'].pivot_table(
index='first_visit_month',
columns='visit_month',
values='uid',
aggfunc='nunique',
)


plt.figure(figsize=(12, 5))
plt.title('Meses de retorno de visitas por cohortes mensuales')
s = sns.heatmap(visits_report_1, annot=True, fmt='.1f', linewidths=1, linecolor='gray',  annot_kws={'size': 9})
sns.set(style='darkgrid')
s.set(xlabel="Month", ylabel="Cohort")

plt.show()
```

<b>Conclusiones: </b> El anterior mapa de calor generado parece indicar que, a partir del mes de septiembre u octubre, la mayor√≠a de los usuarios suele volver en el primer mes desde su primera visita.

### ¬øCu√°ndo empieza la gente a comprar?

```python
orders['buy_date'] = orders['buy_ts'].dt.strftime('%Y.%m.%d')
orders['buy_date'] = pd.to_datetime(orders['buy_date'])
orders.head()
#orders.info()
```

```python
# Obtener la fecha de la primera compra de cada usuario registrado. 
users_first_buy = orders.groupby('uid')['buy_date'].min().reset_index()
users_first_buy
```

```python
first_visit = visits.groupby('uid')['visit_date'].min()
first_visit.name = 'first_visit_date'
first_visit
```

```python
fct_table_1 = users_first_buy.join(first_visit, on='uid')
```

```python
fct_table_1.info()
fct_table_1.head()
```

```python
fct_table_1['days_to_buy'] = fct_table_1['buy_date'] - fct_table_1['first_visit_date']
fct_table_1['days_to_buy'] = fct_table_1['days_to_buy'].dt.days
fct_table_1 = fct_table_1.rename(columns={'buy_date':'first_buy_date'})
fct_table_1
```

```python
fct_table_1[fct_table_1['days_to_buy']<0]
```

```python
fct_table_1[fct_table_1['days_to_buy'] >= 0]
```

```python
fct_table_1['uid'].count()
```

<b>Observaciones: </b> <br>
De los <b>36523</b> usuarios registrados contamos <b>17093</b> cuya primera compra posee una fecha anterior a su primera visita registrada en el a√±o correspondiente a los registros del dataset. Una posible explicaci√≥n podria corresponder al hecho de que los usuarios registrados con esta caracter√≠stica registraron su primera visita al servicio en a√±os anteriores. <br><br>
Con el fin de continuar con un analisis coherente que nos permita identificar el comportamiento de los usuarios, usaremos el segmento de datos correspondiente a aquellos usuarios registrado cuya primera visita sea anterior a la primera compra asociada.

```python
divs = list(range(0,400, 10))
fct_table_1[fct_table_1['days_to_buy'] >= 0]['days_to_buy'].plot(kind='hist',
                                                                 bins=divs,
                                                                 alpha=0.4,
                                                                 color='purple',
                                                                 rot=80, 
                                                                 grid=True, 
                                                                 figsize=(9,5))
plt.xticks(np.arange(0, 400, 10))
plt.show()
```

```python
#Definimos una funci√≥n para calcular la semana de conversion de usuario a cliente.
def calculate_week(x):
    week = np.floor(x/7)
    return round(week)

#Realizamos pruebas
for x in [5, 7, 8, 21, 22, 30]:
    y = calculate_week(x)
    print(y)
```

```python
# Definir nueva columna de conversi√≥n semanal de clientes
fct_table_1['week_conversion'] = fct_table_1['days_to_buy'].apply(calculate_week)
```

```python
# Asignaci√≥n de cohortes mensuales
fct_table_1['first_month'] = fct_table_1['first_visit_date'].astype('datetime64[M]')
fct_table_1['first_month'] = fct_table_1['first_month'].dt.strftime('%Y-%m')
fct_table_1.head()
```

```python
week_conversion_table = fct_table_1[(fct_table_1['days_to_buy']>=0) & (fct_table_1['week_conversion']<=18)].pivot_table(
index='first_month',
columns='week_conversion',
values='uid',
aggfunc='count',
)

sns.set(style='white')
plt.figure(figsize=(14, 5))
plt.title('Semanas transcurridas para primera compra')
sns.heatmap(week_conversion_table, linewidths=1, linecolor='black', annot=True, fmt=".1f", annot_kws={'size': 10})
plt.show()
```

### ¬øCu√°nto dinero traen? (LTV)
<b>Acalaraci√≥n: </b> Idealmente, calcular el LTV requiere conocer la rentabilidad de una empresa. Sin embargo, debido a que entre los datos proporcionados no se especifica esta informaci√≥n y no es posible establecerlo completamente considerando los gastos de marketing, de momento calcularemos usaremos el m√©todo de c√°lculo de LTV considerando unicamente los ingresos generados por ventas. 

```python
# Calcular LVT por cohortes mensuales.
# Primer paso: creaci√≥n de cohortes mensuales
cohorts = orders[orders['buy_date'] >= orders['first_visit_date']].groupby(['first_month_visit','buy_month'])['revenue'].sum()
cohorts = cohorts.reset_index()
cohorts
```

```python
# Calculamos el tama√±o de las cohortes (n√∫mero de usuarios registrados por primera vez en determinado mes)
# Nota: Consideramos dentro del analisis solo a los usuarios cuya primera visita es anterior a su primera compra. 

cohort_sizes = orders[orders['buy_date'] >= orders['first_visit_date']].groupby('first_month_visit')['uid'].nunique()
cohort_sizes = cohort_sizes.reset_index()
cohort_sizes.columns = ['first_month_visit', 'n_buyers']
cohort_sizes
```

```python
report = pd.merge(cohorts, cohort_sizes, on= 'first_month_visit')
report
```

```python
report['buy_month'] = pd.to_datetime(report['buy_month'])
report['first_month_visit'] = pd.to_datetime(report['first_month_visit'])
report['age'] = (report['buy_month'] - report['first_month_visit']) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')
```

```python
report['ltv_1'] = report['revenue'] / report['n_buyers']
report['first_month_visit'] = report['first_month_visit'].dt.strftime('%Y-%m')
report.head()
```

```python
output_1 = report.pivot_table(
index='first_month_visit',
columns='age',
values='revenue',
aggfunc='mean')

output_1.fillna('')


sns.set(style='darkgrid')
plt.figure(figsize=(12, 5))
plt.title('Ganancias mensuales por cohortes')
sns.heatmap(output_1, annot=True, fmt='.1f', linewidths=1, linecolor='black',  annot_kws={'size': 9})
plt.show()
```

```python
output_2 = report.pivot_table(
index='first_month_visit',
columns='buy_month',
values='revenue',
aggfunc='mean')

output_2.fillna('')
sns.set(style='darkgrid')
plt.figure(figsize=(12, 5))
plt.title('Ganancias mensuales por cohortes')
sns.heatmap(output_2, annot=True, fmt='.1f', linewidths=1, linecolor='black',  annot_kws={'size': 9})
plt.show()
```

```python
output_3 = report.pivot_table(
index='first_month_visit',
columns='age',
values='ltv_1',
aggfunc='mean')


sns.set(style='darkgrid')
plt.figure(figsize=(12, 5))
plt.title('LTV por cohortes')
sns.heatmap(output_3, annot=True, fmt='.1f', linewidths=1, linecolor='black',  annot_kws={'size': 9})
plt.show()
```

```python
ltv_cohorts = report.groupby('first_month_visit')['ltv_1'].sum().reset_index()
ltv_cohorts
```
